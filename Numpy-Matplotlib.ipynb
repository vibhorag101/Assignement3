```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)             # 1 row and 2 columns, with a plot ID of 1
plt.plot(x,y, marker = 'o')
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)             # 1 row and 2 columns, with a plot ID of 2
plt.plot(x,y, marker = 'o')
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)             # 2 rows and 1 column, with a plot ID of 1
plt.plot(x,y, marker = 'o')
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)             # 2 rows and 1 column, with a plot ID of 2
plt.plot(x,y, marker = 'o')
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.tight_layout()               # adds padding to the subplots
plt.show()
```

```python
import matplotlib.pyplot as plt

circle = plt.Circle((0, 0), 1.5, color='g', fill=False,linestyle="dashed")
fig, ax = plt.subplots()  # returns figure layout and an array of axes for each subplot. Number of rows and columns can be specified as well.

ax.add_patch(circle)
ax.set(xlim=(-2, 2), ylim = (-2, 2))
ax.set_aspect(1)   # sets the height (y-unit) to width (x-unit) ratio
plt.show()
```

```python
import matplotlib.pyplot as plt

rectangle = plt.Rectangle((-2, -2), 4, 3, color='r', fill=False)
fig, ax = plt.subplots()   # returns figure layout and an array of axes for each subplot
ax.add_patch(rectangle)
ax.set(xlim=(-5, 5), ylim = (-5, 5))
ax.set_aspect(1)   # sets the height (y-unit) to width (x-unit) ratio
plt.show()
```
